```python
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences=tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences=pad_sequences(sequences,padding='post')
sequences
```

```python
from keras.models import Sequential
from keras.layers import Embedding

model=Sequential()
model.add(Embedding(17,2,input_length=5))
model.summary()
```

```python
model.compile('adam','accuracy')
pred=model.predict(sequences)
print(pred)
```

```python
from keras.datasets import imdb
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(x_train,y_train),(x_test,y_test)=imdb.load_data()
```

```python
maxlen = max(len(seq) for seq in x_train)
x_train = pad_sequences(x_train, padding='post', maxlen=maxlen)
x_test = pad_sequences(x_test, padding='post', maxlen=maxlen)
```

```python
print(maxlen)
```

```python
x_train.shape
```

```python
#find the size of the vocab
import numpy as np
size=len(np.unique(x_train)) + 1  # Add 1 to account for zero-based indexing
size
```

```python
model=Sequential()
model.add(Embedding(size, 2, input_length=2494))  # Corrected order of arguments
model.add(SimpleRNN(32,input_shape=(maxlen,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
history=model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
# Plot Train and test accuracy form history
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
```

```python
# Plot val loss graph
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
```
