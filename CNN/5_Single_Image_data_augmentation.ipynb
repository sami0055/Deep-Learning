```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
img=load_img('/content/cat.jpg',target_size=(256,256))
```

```python
img
```

```python
datagen=ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
img=img_to_array(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,256,256,3)
```

```python
# Image generate
i=0
for batch in datagen.flow(input_batch,batch_size=1,save_to_dir='preview',save_prefix='cat',save_format='jpeg'):
  i+=1
  if i>10:
    break
```
